In the next problem, we will use decision trees to predict a data set used for Kaggle competitions in older iterations of the course; see e.g.,

First, let's load the data:

```python
url = 'https://www.ics.uci.edu/~ihler/classes/cs178/data/precip_train.txt'

with requests.get(url, verify=False) as link: data = np.genfromtxt(StringIO(link.text),delimiter=None)

X,Y = data[:,:-1], data[:,-1]
```

#### Problem 2.1
Compute and print some useful statistics about the data -- the minimum, maximum, mean, and variance of each of the features.

```python
stats = {
    "Minimum": np.min(X, axis=0),
    "Maximum": np.max(X, axis=0),
    "Mean": np.mean(X, axis=0),
    "Variance": np.var(X, axis=0)
}
print(stats)
```

#### Problem 2.2
To allow faster experimentation, select the first 10,000 data points as training data, ``Xtr, Ytr``, and the next 10,000 data points as validation, ``Xva, Yva``.  Then, learn a decision tree classifier on ``Xtr`` using the ``DecisionTreeClassifier`` class from ``scikit-learn``.  Use a large depth, say ``max_depth=100``, and the Shannon entropy impurity function in your training.  Also, use ``random_state=seed`` to ensure consistency.

Compute and print out the training error rate and validation error rate of your model.

```python
from sklearn.tree import DecisionTreeClassifier

# Splitting the data
Xtr, Ytr = X[:10000], Y[:10000]
Xva, Yva = X[10000:20000], Y[10000:20000]

# Training the decision tree classifier
DTC = DecisionTreeClassifier(max_depth=100, criterion='entropy', random_state=seed)
DTC.fit(Xtr, Ytr)

# Computing error rates
train_error_rate = 1 - DTC.score(Xtr, Ytr)
val_error_rate = 1 - DTC.score(Xva, Yva)

print(f"Training Error Rate: {train_error_rate}")
print(f"Validation Error Rate: {val_error_rate}")
```

#### Problem 2.3
Now try varying the ``max_depth`` parameter, , which forces the tree learning algorithm to stop after at most that many levels.  Test ``max_depth`` values in the range {1, 2, 3, ..., 16}, and plot the training and validation error rates versus ``max_depth``.
Do models with higher ``max_depth`` have higher or lower complexity?
What choice of ``max_depth`` provides the best decision tree model?

```python
# store error rates
train_errors = []
val_errors = []

# Iterate over max_depth values
for depth in range(1, 17):
    DTC2 = DecisionTreeClassifier(max_depth=depth, criterion='entropy', random_state=seed)
    DTC2.fit(Xtr, Ytr)
    
    # Compute error rates
    train_error = 1 - DTC2.score(Xtr, Ytr)
    val_error = 1 - DTC2.score(Xva, Yva)
    
    # Store the error rates
    train_errors.append(train_error)
    val_errors.append(val_error)

# Plot the error rates
plt.figure(figsize=(10, 6))
plt.plot(range(1, 17), train_errors, label='Training Error')
plt.plot(range(1, 17), val_errors, label='Validation Error')
plt.xlabel('Max Depth')
plt.ylabel('Error Rate')
plt.title('Error Rates vs Max Depth for Decision Tree')
plt.legend()
plt.show()
```

Models with higher max_depth are more complex because they have more decision nodes and can create more intricate decision boundaries. The best max_depth is the one that provides the lowest validation error, indicating the model that generalizes best. In this case, that's around 4.

#### Problem 2.4
The ``min_samples_leaf`` parameter  controls the complexity of decision trees by lower bounding the amount of data required to split nodes when learning.  Fixing ``max_depth=100``, compute and plot the training and validation error rates for ``min_samples_leaf`` values in the range
``{2**0, 2**1, 2**2, ... 2**17}``. 
Do models with higher ``min_samples_leaf`` have higher or lower complexity?
What choice of ``min_samples_leaf`` provides the best decision tree model?
Is this model better or worse than your depth-controlled model?

```python
# store error rates
train_errors2 = []
val_errors2 = []

# Iterate over min_samples_leaf values as powers of 2
for i in range(18):
    msl = 2**i
    DTC3 = DecisionTreeClassifier(max_depth=100, min_samples_leaf=msl, criterion='entropy', random_state=seed)
    DTC3.fit(Xtr, Ytr)
    
    # Compute error rates
    train_error = 1 - DTC3.score(Xtr, Ytr)
    val_error = 1 - DTC3.score(Xva, Yva)
    
    # Store the error rates
    train_errors2.append(train_error)
    val_errors2.append(val_error)

# Plotting the error rates
plt.figure(figsize=(10, 6))
plt.plot([2**i for i in range(18)], train_errors2, label='Training Error')
plt.plot([2**i for i in range(18)], val_errors2, label='Validation Error')
plt.xlabel('Min Samples Leaf')
plt.ylabel('Error Rate')
plt.xscale('log', base=2)  # Setting x-axis to logarithmic scale for better visualization
plt.title('Error Rates vs Min Samples Leaf for Decision Tree')
plt.legend()
plt.show()
```

Models with higher min_samples_leaf have lower complexity. This is because they require more data in each leaf, leading to fewer splits and a simpler model structure. The optimal min_samples_leaf is the one that yields the lowest validation error, indicating the best generalization performance. In this case, that's 2^8. This model is better than the depth-controlled model.

## Problem 3: Random Forests

Although ``scikit`` has its own implementations of bagging estimators and random forests, here we will build a random forest model manually, for illustration.

Random Forests are bagged collections of decision trees, which select their decision nodes from randomly chosen subsets of the possible features (rather than all features).  You can implement this easily in ``DecisionTreeClassifier`` using option ``max_features=n``, where ``n`` is the number of features to select from (e.g., n=50 or n=60 might be good choices if there are 90-some features).
You'll write a for-loop to build the ensemble members, and another to compute the prediction of the ensemble.

#### Problem 3.1 

Using your validation split, learn a bagged ensemble of decision trees on the training data
and evaluate validation performance.
(See the pseudocode from lecture slides.)
For your individual learners, use little complexity control (depth cutoff 15+, min leaf 1-4, etc.), since the bagging will be used to control overfitting instead.

You will implement bootstrap sampling yourself (again, see pseudocode); simply generate $m'$ data indices uniformly with replacement, using ``numpy`` functions or simply ``rand`` and conversion to integers.
For your bootstrap process, draw the same number of data as in your training set after the
validation split (i.e., set $m' = m$, the number of training data).

Learn at least 50 ensemble members, as you will use them in the next part.

```python
from scipy.stats import mode

# Number of trees in the random forest
n_trees = 50

# List to store each tree in the ensemble
forest = []

# Training each tree in the forest
for _ in range(n_trees):
    # Generate bootstrap sample
    bootstrap_indices = np.random.randint(0, len(Xtr), len(Xtr))
    X_sample, Y_sample = Xtr[bootstrap_indices], Ytr[bootstrap_indices]

    # Train decision tree on bootstrap sample
    tree = DecisionTreeClassifier(max_depth=30, min_samples_leaf=2, max_features=50, random_state=seed)
    tree.fit(X_sample, Y_sample)

    # Add the trained tree to the forest
    forest.append(tree)

# Predict with the forest on the validation data
predictions = np.array([tree.predict(Xva) for tree in forest])

# Aggregate predictions: majority vote
ensemble_predictions = mode(predictions, axis=0)[0][0]

# Calculate validation error
validation_error = np.mean(ensemble_predictions != Yva)

print(f"Validation Error of the Random Forest: {validation_error}")
```

#### Problem 3.2 

Plot the training and validation error of the ensemble as a function of the number of learners $B$ that you include in the ensemble, for (at least) values $B \in \{1,5,10,25,50\}$.  (Just use the first $B$ learners that you trained in part 1.)

Remember that your ensemble predicts by making a prediction for each member model, and then taking a vote among the $B$ models.  Since your target classes are +1 and -1, as in the slide pseudocode, to find the outcome of the vote you can simply check whether the average of the class predictions is positive or negative.

```python
# Defining B values
B_values = [1, 5, 10, 25, 50]

# Store errors for each value of B
train_errors3 = []
val_errors3 = []

# Evaluate ensemble
for B in B_values:
    # Use the first B learners for prediction
    predict_train = np.array([tree.predict(Xtr) for tree in forest[:B]])
    predict_val = np.array([tree.predict(Xva) for tree in forest[:B]])

    # Aggregate predictions by checking if average prediction is positive or negative
    ensemble_predict_train = np.sign(predict_train.mean(axis=0))
    ensemble_predict_train[ensemble_predict_train == 0] = 1  # set to 1 if mean = 0
    # print(ensemble_predict_train)
    ensemble_predict_val = np.sign(predict_val.mean(axis=0))
    ensemble_predict_val[ensemble_predict_val == 0] = 1  # set to 1 if mean = 0
    # print(ensemble_predict_val)

    # Compute errors
    train_error = np.mean(ensemble_predict_train != Ytr)
    val_error = np.mean(ensemble_predict_val != Yva)

    # Store the errors
    train_errors3.append(train_error)
    val_errors3.append(val_error)

# Plotting the error rates
plt.figure(figsize=(10, 6))
plt.plot(B_values, train_errors3, label='Training Error')
plt.plot(B_values, val_errors3, label='Validation Error')
plt.xlabel('Number of Learners (B)')
plt.ylabel('Error Rate')
plt.title('Training and Validation Errors vs. Number of Learners')
plt.legend()
plt.show()
```
